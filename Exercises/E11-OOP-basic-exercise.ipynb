## 1. Step tracker

Create a class that can be used as a step tracker. It should have a property "steps" that is read only; a method step() that increases "steps" by 1 each time it is called; and a method reset() that resets the counter.

Instantiate the class, and write a loop that simulates walking 1000 steps. Then print the value of "steps".

```python
class Steps:

    def __init__(self, steps):
        self._steps = steps

    def reset(self):
        self._steps = 0 

    def step(self):
        self._steps += 1
    
    @property
    def steps(self):
        return self._steps


my = Steps(0)

for ss in range(1000):
    my.step()

my.reset()

print(f"Steg: {my.steps}")


    
#print(my.steps)
#print(my.steps)


```

## 3. Red and blue (*)

Create a class that has a property "red", and a property "blue". Both should be floats, and be able to take any value between 0.0 and 100.0. However, they should be "linked" in such a way that the sum of "red" and "blue" always is 100.0. i.e. if we set the value of "blue" to 8.5, and then read the value of "red", it should return 91.5

```python
class Colors:

    def __init__(self, blue = 50.0, red = 50.0):
        self._blue = blue
        self._red = red


    @property
    def blue(self):
        return self._blue

    @blue.setter
    def blue(self, blue):
        self._blue = f"Blue: {blue}" if 0.1 <= blue <= 99.9 else "Sorry, blue must be between 0.1 and 99.9"
        self._red = f"Red: {100-blue}" if 0.1 <= blue <= 99.9 else "Red remains unchanged"

    @property
    def red(self):
        return self._red

    @red.setter
    def red(self, red):
        self._red = f"Red: {red}" if 0.1 <= red <= 99.9 else "Sorry, red must be between 0.1 and 99.9"
        self._blue = f"Blue: {100-red}" if 0.1 <= red <= 99.9 else "Blue remains unchanged"


color = Colors()

color.blue = 43.6

print(color.blue)
print(color.red)
```

## 4. One thousand cars (*)

Create a class that represents a car. Every car can have a color and a length. When a new car is instantiated it gets a random color, and a random length (between 3 and 5 meters). Instatiate 1000 cars a save them in a list. Then print the sum of the length of all green cars in the list.

```python
import random
COLORS = ['velvet red', 'cobalt blue', 'royal purple', 'emerald green', 'obsidian black']

class Car:

    def __init__(self):
        self.color = random.choice(COLORS)
        self.length = random.randint(3,5)
        

thousand_cars = [Car() for n in range(1000)]


length_of_greens = 0
for car in thousand_cars:
    length_of_greens += car.length if car.color == 'emerald green' else 0


print(length_of_greens)    


```

## 5. Email (**)

Create an email class with a property "address". When we set this property it should validate that we gave it a proper email address by checking that it contains one or more letter, followed by an at-sign (@), followed by one or more letter, then a dot (.), then at least to letters.

We should also be able to provide the address directly, when creating an new instance of the class (it must still be validated).

The class should also have the following properties: "username", "domainname", and "topdomain" implemented in such a way that, if we set the address to "fredrik@everyloop.com", the username should read "fredrik", the domainname should read "everyloop.com", and the topdomain should read "com".

When changing any of the four properties, all the others must be updated accordingly; and the address must always remain valid.

```python
class Email:

    def __init__(self, address):
        self.address = address


    @property
    def address(self):
        return self._address

    @address.setter
    def address(self, address):

        at_index = address.index('@')
        dot_index = address[::-1].index('.')

        if '@' in address[1:] and '.' in address[at_index+2:] and '.' in address[at_index+2:-dot_index+1]: 

            username = address[:at_index]
            domainname = address[at_index+1:]
            topdomain = address[-dot_index:]

            self._address = f"{username}@{domainname[:-len(topdomain)]}{topdomain}"
            
            self._username = username
            self._domainname = domainname
            self._topdomain = topdomain

        else:
            print("Sorry, not valid address")
            self._address = self.address

    ####################################

    @property
    def username(self):
        return self._username
    
    @username.setter
    def username(self, username):
        original_username_length = len(self.username)
        self._username = username if len(username) >= 1 else self.username

        self._address = f"{self.username}{self.address[original_username_length:]}"

    ####################################

    @property
    def domainname(self):
        return self._domainname
    
    @domainname.setter
    def domainname(self, domainname):
        original_domainname_length = len(self.domainname)
        dot_index = domainname.index(".")
        
        if "." in domainname[1:] and "." in domainname[:-2]:
            self._domainname = domainname  
            self._topdomain = domainname[dot_index + 1:]
            self._address = f"{self.address[:-original_domainname_length]}{self.domainname}"

        else:
            self._domainname = self.domainname

    ####################################

    @property
    def topdomain(self):
        return self._topdomain
    
    @topdomain.setter
    def topdomain(self, topdomain):
        original_topdomain_length = len(self.topdomain)
        
        if len(topdomain) > 1:
            self._topdomain = topdomain  
            self._domainname = f"{self.domainname[:-original_topdomain_length]}{self.topdomain}"
            self._address = f"{self.address[:-original_topdomain_length]}{self.topdomain}"

        else:
            self._topdomain = self.topdomain
            self._domainname = self.domainname



mail = Email("ale.mar.vill@gmail.com")



mail.username
#mail.domainname = 'hotm.se'
#mail.topdomain = "es"

#mail.address

```

```python
from dataclasses import dataclass


@dataclass
class Employee:
    
    raise_pct = 1.06

    first: str
    last: str
    salary: int


    def profile(self):
        print(self.first, self.last, self.salary)

    def salary_raise(self):
        print(self.salary*self.raise_pct)




@dataclass
class Developer(Employee):

    raise_pct = 1.04

    language: str
    years_of_experience: int

    def profile(self):
        print(self.first, self.last, self.salary*self.raise_pct, self.language, self.years_of_experience)




emp = Employee("Ale", "Vill", 45000)
dev = Developer("Alejandro", "Martinez", 40000, "Python", 4)
emp.profile()
emp.salary_raise()

dev.profile()
dev.salary_raise()



```
