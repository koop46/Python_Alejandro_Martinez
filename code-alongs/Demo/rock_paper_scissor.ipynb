```python
from random import choice


SIGNS = ['rock', 'paper', 'scissor']


def main():

    #rock, paper, scissors

    rounds = welcome()

    for round in range(1, rounds + 1):
        game(round)


############################################################################################################


def welcome():

    print(f"Welcome to the {', '.join(SIGNS)} game \n")
    print("Rules of the game: ")

    for win, loss in zip([0,1,2],[2,0,1]):
        print(f"- {SIGNS[win].title()} beats {SIGNS[loss]}.")

    print("")
    print("How may rounds you wanna go, 3 or 5?")

    while True:
        rounds = int(input("3 or 5 rounds?"))

        if rounds == 3 or rounds == 5:
            return rounds
        else:
            print("Sorry, must be 3 or 5")


def game(round):

    print(f"\nRound {round}, fight!")
    player = player_sign()
    computer = computer_sign()
    fight(player, computer)


def player_sign():

    print("- Rock, paper or scissor?")
    player_choice = input("- Rock, paper or scissor?")
    print(f"You chose {player_choice}")
    
    return player_choice


def computer_sign():

    computer_choice = choice(SIGNS)
    print(f"Computer chose: {computer_choice}")

    return computer_choice


def fight(player_sign, computer_sign):

    print("")
    for winner, loser in zip([0,1,2],[2,0,1]):
        if player_sign == SIGNS[winner] and computer_sign == SIGNS[loser]:
            print(f"\n >>> Player wins by {player_sign}!")
            return 

        elif computer_sign == SIGNS[winner] and player_sign == SIGNS[loser]:
            print(f"\n >>> Computer wins by {computer_sign}!")
            return 

        else:
            print( "It's a tie!")
            return 







main()
```
